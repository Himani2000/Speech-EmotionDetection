```python
#This consists of all the 4-datasets used
main_data='All Datasets'
files=[]
for file in os.listdir(main_data):
    files.append(main_data+'/'+file)
files
```

```python
emotion_type=[]
file_path=[]
for emo_file in os.listdir(files[3]):
    
    if emo_file[-8:-6]=='_a':
        emotion_type.append('male_angry')
        
    elif emo_file[-8:-6]=='_d':
        emotion_type.append('male_disgust')
        
    elif emo_file[-8:-6]=='_f':
        emotion_type.append('male_fear')
        
    elif emo_file[-8:-6]=='_h':
        emotion_type.append('male_happy')
        
    elif emo_file[-8:-6]=='_n':
        emotion_type.append('male_neutral')
        
    elif emo_file[-8:-6]=='sa':
        emotion_type.append('male_sad')
        
    elif emo_file[-8:-6]=='su':
        emotion_type.append('male_surprise')
    
    file_path.append(files[3]+'/'+emo_file)
    
    
        
```

```python
df_savee=pd.DataFrame({'Emotion':emotion_type,'Path':file_path})
df_savee['Emotion'].value_counts()
```

```python
audio_path=[]
gender=[]
emotion=[]
for file in os.listdir(files[2]):
    path=os.path.join(files[2],file)
    path_files=os.listdir(path)
    for audio_file in path_files:
        name=audio_file.split('.')[0].split('-')
        emotion.append(int(name[2]))
        
        if(int(name[6])%2==0):
            gender.append('female')
        
        else:
            gender.append('male')
        
        audio_path.append(os.path.join(os.path.join(files[2],file),audio_file))
```

# FEATURE GENERATION

```python
#Feature extraction 
```

```python
features=[]
```

```python
def toFeature(file_paths):
    features=[]
    for path in file_paths:
    
        X, sample_rate = librosa.load(path, res_type='kaiser_fast',duration=2.5,sr=44100,offset=0.5)
        sample_rate = np.array(sample_rate)
        mfccs = np.mean(librosa.feature.mfcc(y=X,sr=sample_rate, n_mfcc=13),axis=0)
        features.append([mfccs])
    return features
```

```python
file_paths=list(main_df['Path'])
features_list=toFeature(file_paths)
len(features_list)
```

```python
len(features_list[1][0])
```

```python
df_features=pd.DataFrame({'feature':features_list})
```

```python
df_final=pd.concat([main_df,pd.DataFrame(f['feature'].values.tolist())],axis=1)
```

```python
import pandas as pd 
import numpy as np
import os 
```

```python
dataset='ravdess-emotional-speech-audio'
os.listdir(dataset)
```

```python
stri='03-01-06-02-02-01-02.wav'
stri.split('.')[0].split('-')
```

```python
audio_path=[]
gender=[]
emotion=[]
for file in os.listdir(dataset):
    path=os.path.join(dataset,file)
    path_files=os.listdir(path)
    for audio_file in path_files:
        name=audio_file.split('.')[0].split('-')
        emotion.append(int(name[2]))
        
        if(int(name[6])%2==0):
            gender.append('female')
        
        else:
            gender.append('male')
        
        audio_path.append(os.path.join(os.path.join(dataset,file),audio_file))
```

```python
len(gender),len(emotion),len(audio_path)
```

```python
mapping={1:'neutral', 2:'neutral', 3:'happy', 4:'sad', 5:'angry', 6:'fear', 7:'disgust', 8:'surprise'}
df=pd.DataFrame({'gender':gender,'emotion':emotion})
df=df.replace({'emotion':mapping})
df.head()
```

```python
df['labels']=df['gender']+'_'+df['emotion']
df
```

```python
df['labels'].value_counts()
```

```python
df['audio_path']=audio_path
df.drop(['emotion','gender'],axis=1,inplace=True)
df
```

```python
df.to_csv('RAVDESS_LabelPath.csv',index=False)
```
